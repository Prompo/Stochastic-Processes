ΑΠΑΝΤΗΣΕΙΣ:


i)Elbow method: Είναι μία τεχνική που χρησιμοποιείται για να βοηθήσει την επιλογή του βέλτιστου αριθμού των clusters σε ένα σύνολο δεδομένων κατά την εκτέλεση του αλγοιρίθμου K-Means.
Κατά την εκτέλεση του K-means με διάφορους αριθμούς clusters, η Inertia(Sum of Squared Distances) μειώνεται καθώς ο αριθμός των clusters αυξάνεται. Ωστόσο, σε κάποιο σημείο, η μείωση αυτή επιβραδύνεται και η γραφική παράσταση της Inertia σε σχέση με τον αριθμό των clusters μοιάζει με έναν αγκώνα(elbow). Το σημείο όπου ο "αγκώνας" εμφανίζεται είναι συνήθως ένα καλό σημείο επιλογής για τον βέλτιστο αριθμό clusters.
Συγκεκριμένα, το "Elbow Method" προτείνει τον αριθό των clusters στον οποίο η μείωση της Inertia αρχίζει να επιβραδύνετε σημαντικά, δηλαδή το σημείο όπου η καμπύλη της Inertia παίρνει τη μορφή ενός αγκώνα. Σε αυτό το σημείο , η προσθήκη επιπλέον clusters δεν θα προσφέρει σημαντική μείωση της Inertia.
Στα παρακάτω διαγράμματα παρατηρούμε ότι οι καλύτερες επιλογές για αριθμό clusters είναι: cluster=3 και cluster=4. Σε αυτά τα σημεία δημιουργείται ο αγκώνας. Ωστόσο, το σημείο όπου η μείωση επιβραδύνεται ολοκληρωτικά είναι το σημείο για cluster=4.   

ii) Αλγόριθμος K-Means:
Ο αλγόριθμος K-Means είναι ένας αλγόριθμος συσταδοποίησης (clustering) που χρησιμοποιείται για να ομαδοποιήσει τα δεδομένα σε κατηγορίες ή clusters. Η βασική ιδέα είναι να χωρίσει τα δεδομένα σε ομάδες όπου τα μέλη κάθε ομάδας είναι κοντά μεταξύ τους και οι ομάδες είναι όσο το δυνατόν πιο μακριά μεταξύ τους.
```
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm
from sklearn.datasets import make_blobs
import seaborn as sns

import pandas as pd
import numpy as np
```

```
# import some data to play with (from sklearn)
iris = datasets.load_iris()
print(iris)
```

```
#print is optional
print(iris.data)
iris.feature_names
print(iris.target)
iris.target_names
```

```
# Store the inputs as a Pandas Dataframe and set the column names
x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
```

```
from yellowbrick.cluster import KElbowVisualizer
import matplotlib
from matplotlib.colors import LinearSegmentedColormap
import itertools
from mpl_toolkits.axes_grid1 import make_axes_locatable

y = pd.DataFrame(iris.target)
y.columns = ['Targets']

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
lista=[]
for k in range(2,13):
  model=KMeans(n_clusters=k,n_init=10,random_state=42)
  model.fit(x)
  lista.append(model.inertia_)
  print(f"Για k={k}, η inertia ειναι: {model.inertia_:.3f}")

plt.plot(range(2, 13), lista, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters ')
plt.ylabel('Within-Cluster Sum of Squares ')
plt.show()

#The cluster in which appears the elbow in the diagram below is k=4.
#So, n_cluster=4
kmean=KMeans(n_clusters=4,n_init=10,random_state=42)
kmean.fit(x)
predictions=kmean.predict(x)
print(f"\npredictions:\n{predictions}")

matplotlib.rc_file_defaults()

accuracy = sm.accuracy_score(y, predictions)
print(f'Η ακρίβεια του μοντέλου συσταδοποίησης ισούται με {100*accuracy:.3f}%.')
print('Η μήτρα σφαλμάτων απεικονίζεται ακολούθως:')
confusion_matrix = sm.confusion_matrix(y, predictions)
print('Confusion Matrix:')
print(confusion_matrix)
print("\n")



model = KMeans(n_init=10,random_state=42)
KElbowVisualizer(model, k=(2,13)).fit(x).show()
print()
```
