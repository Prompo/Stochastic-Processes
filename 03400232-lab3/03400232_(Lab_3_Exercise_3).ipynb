ΑΠΑΝΤΗΣΕΙΣ:

i) Από τα δεδομένα της εκφώνησης έχουμε ότι η Μαρκοβιανή αλυσίδα αποτελείται από το σύνολο καταστάσεων Χ={1,2,3,4,5,6,7,8}. Από θεωρία, γνωρίζουμε ότι:

Ανοικτές κλάσεις είναι αυτές που επιτρέπουν έξοδο προς άλλη κλάση.

Κλειστές αυτές που δεν επιτρέπουν έξοδο και οι στατιστικές ιδιότητες της αλυσίδας μετά την παρέλευση μεταβατικού χρόνου περιορίζονται στο υποσύνολο αυτό.

Επίσης, αν υπάρχει μόνο μία κλειστή κλάση η διαδικασία είναι Irreducible.

Στη συγκεκριμένη άσκηση, η αλυσίδα ανάγεται σε τρεις κλάσεις επικοινωνίας: τις
{1,2} και {6,7,8}, οι οποίες είναι κλειστές, και την {3,4,5}, η οποία είναι ανοιχτή. Πρόσθετα, η κλάση {3,4,5} είναι παροδική, ως ανοιχτή, ενώ οι κλάσεις
{1,2} και {6,7,8} είναι επαναληπτικές, ως κλειστές και πεπερασμένες.

Τη συγκεκριμένη άποψη, μπορούμε να την αναλύσουμε λίγο περισσότερο. Παρατηρούμε στη μήτρα πιθνοτήτων ότι η κατάσταση 1 μπορεί να μεταβεί με ισόποση πιθανότητα είτε στον εαυτό της είτε στην κατάσταση 2. Το ίδιο ισχύει για τη κατάσταση 2 όσον αφορά τις εναλλακτικές επιλογές. Συγκεκριμένα, η κατάσταση 2 μπορεί να μεταβεί με πιθανότητα 1/4 στη κατάσταση 1 και να ξαναμεταβεί στον εαυτό της με πιθανότητα 3/4. Συνεπώς,οι καταστάσεις 1 και 2 μπορούν να μεταβούν μόνο μεταξύ τους. Γι΄αυτό το λόγο, η κλάση {1,2} είναι κλειστή.

Η άλλη κλάση που είναι επίσης κλειστή, είναι η {6,7,8}. Ο λόγος είναι ίδιος με πριν. Οι καταστάσεις 6,7,8 μπορούν να μεταβούν μόνο μεταξύ τους και σε καμία άλλη κατάσταση του χώρου.Η κατάσταση 6 μπορεί να μεταβεί με ισόποση πιθανότητα είτε στη κατάσταση 7 είτε στη κατάσταση 8, ενώ δεν υπάρχει καμία πιθανότητα να ξαναμεταβεί στον εαυτό της. Η κατάσταση 7 μπορεί να μεταβεί με ισόποση πιθανότητα είτε στη κατάσταση 6 είτε στη κατάσταση 8, ενώ δεν υπάρχει καμία πιθανότητα να μεταβεί ξανά στον εαυτό της. Τέλος, η κατάσταση 8 μπορεί να μεταβεί με ισόποση πιθανότητα είτε στη κατάσταση 6 είτε στη κατάσταση 7, ένω ,όπως και στις προηγούμενες περιπτώσεις, δεν υπάρχει περίπτωση να μεταβεί ξανά στον εαυτό της.

Η κλάση {3,4,5} είναι ανοιχτή γιατί περιλαμβάνει καταστάσεις που μπορούν να μεταβούν σε κατστάσεις που ανήκουν σε άλλες κλάσεις. Συγκεκριμένα, η κατάσταση 3 μπορεί να μεταβεί στις καταστάσεις 1 και 2. Επίσης, η κατάσταση 5 μπορεί να μεταβεί στις καταστάσεις 6 και 7. Προφανώς, οι καταστάσεις 3,4,5 μπορούν να μεταβούν όλες μεταξύ τους.

ii) Χρησιμοποιώντας το αποτέλεσμα του προηγούμενου ερωτήματος, προσδιορίζουμε την αναλλοίωτη κατανομή της μαρκοβιανής αλυσίδας για την περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση (α) 1 και (β) 8, χωρίς τη βοήθεια υπολογιστή.

α) Στην περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση 1, τότε υπάρχουν μόνο 2 πιθανές μετακινήσεις.Η αλυσίδα να παραμείνει στην κατάσταση 1 με πιθανότητα 0.5 ή να μεταβεί στην κατάσταση 2 με πιθανότητα 0.5. Από τη κατασταση 2, η αλυσίδα παραμένει στο 2 με πιθανότητα 0.75 ή επιστρέφει στο 1 με πιθανότητα 0.25. Μη ξεχνάμε άλλωστε ότι οι κατάστασεις 1 και 2 ανήκουν στο κλειστό σύνολο {1,2}. Άρα, η μήτρα πιθανοτήτων είναι:

(π1,π2)=(π1,π2)*Ρ , όπου π1+π2=1, και Ρ η μήτρα πιθανοτήτων

πΑ=(1/3,2/3,0,0,0,0)
